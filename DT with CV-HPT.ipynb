```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import balanced_accuracy_score
```

```python
df=pd.read_csv('data2.csv')
```

```python
x = df.drop(columns=['severity_class'])
y = df[['severity_class']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=False)
```

```python
clf = DecisionTreeClassifier(max_depth=4, min_samples_split=5) 
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
balance = balanced_accuracy_score(y_test, y_pred)
print(f"Balanced Accuracy: {balance}")
```

```python
from sklearn.model_selection import KFold, GridSearchCV
```

```python
k = 5
fold_size = len(x) // k
for i in range(k):
    x_test2 = x.iloc[i * fold_size : (i + 1) * fold_size]
    y_test2= y.iloc[i * fold_size : (i + 1) * fold_size]
    x_train2 = pd.concat([x.iloc[:i * fold_size], x.iloc[(i + 1) * fold_size:]])
    y_train2 = pd.concat([y.iloc[:i * fold_size], y.iloc[(i + 1) * fold_size:]])
```

```python
dectree = DecisionTreeClassifier(
    criterion='gini',      # Criterion: 'gini' (default) or 'entropy' to measure split quality
    max_depth=None,        # Max depth of the tree. Use None for no limit.
    min_samples_split=2,   # Minimum samples needed to split an internal node
    min_samples_leaf=1,    # Minimum samples a leaf node must have   
    
)


dectree.fit(x_train2, y_train2)
y_pred2 = dectree.predict(x_test2)

balance = balanced_accuracy_score(y_test2, y_pred2)
print(balance*100)
```

```python
cv = KFold(n_splits = 5, shuffle=False)  ### Specifies the k=5, number of splits in the cross-validation procedure

logReg = DecisionTreeClassifier() ### Specify the model

params = {
    'criterion': ['gini', 'entropy'],  
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 2, 4], 
    'max_depth':[None, 10, 20, 30]
}

grid_search = GridSearchCV(
    estimator=dectree,       # Model
    param_grid=params,      # Parameter Grid: 
    scoring = 'balanced_accuracy',
    cv=cv,
    verbose=2,              # Print progression: 1 = very brief, 2 = detailed, 0 = none
    n_jobs=4,               # Number of processors to use from my CPU
)

grid_search.fit(x, y)
```

```python
print()
best_params = grid_search.best_params_
for param, value in best_params.items():
    print(f"{param}: {value}"+',  ')
```
