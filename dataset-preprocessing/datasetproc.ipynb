```python
import numpy as np
import pandas as pd
```

```python
path_to_data = 'NTSB avall file dataset_1.xlsx'
ntsb=pd.ExcelFile(path_to_data)
ntsb1 = pd.read_excel(ntsb, 'Aircraft related')
ntsb2 = pd.read_excel(ntsb, 'Event related')
ntsb3 = pd.read_excel(ntsb, 'Findings')
ntsb4 = pd.read_excel(ntsb, 'Engines')
ntsb5 = pd.read_excel(ntsb, 'Flight_Crew')
```

```python
print(len(ntsb1))
print(len(ntsb2))
print(len(ntsb3))
print(len(ntsb4))
print(len(ntsb5))
```

```python
ntsb1 = pd.read_excel(path_to_data, sheet_name='Aircraft related', usecols=['ev_id','cert_max_gr_wt','acft_category','homebuilt','afm_hrs_last_insp','damage'])
ntsb2 = pd.read_excel(path_to_data, sheet_name='Event related', usecols=['ev_id','ev_dow', 'ev_time','ev_year', 'ev_month','ev_nr_apt_loc',
       'apt_dist','apt_elev','light_cond', 'sky_cond_nonceil','altimeter',
       'sky_nonceil_ht', 'sky_ceil_ht', 'sky_cond_ceil', 'vis_sm', 'wx_temp', 'wx_dew_pt', 'wind_dir_deg','wind_vel_kts', 'dec_latitude', 'dec_longitude', 'gust_kts','ev_highest_injury','wx_cond_basic','dec_latitude', 'dec_longitude'])
ntsb3 = pd.read_excel(path_to_data, sheet_name='Findings', usecols=['ev_id','Cause_Factor'])
ntsb4 = pd.read_excel(path_to_data, sheet_name='Engines', usecols=['ev_id','eng_type'])
ntsb5 = pd.read_excel(path_to_data, sheet_name='Flight_Crew', usecols=['ev_id','crew_age','med_certf','med_crtf_vldty','crew_tox_perf','crew_no'])
```

```python
dataframes = [ntsb1, ntsb2, ntsb3, ntsb4, ntsb5]
merged_df = dataframes[0]
for df in dataframes[1:]:
    merged_df = pd.merge(merged_df, df, on='ev_id', how='inner')
print(merged_df)
```

```python
filtered_df = merged_df[(merged_df['acft_category'] == "AIR")]
print(filtered_df)
```

```python
output_filename = 'dataset2.csv'
filtered_df.to_csv(output_filename, index=False)
print(f"Data saved to {output_filename}")
```
