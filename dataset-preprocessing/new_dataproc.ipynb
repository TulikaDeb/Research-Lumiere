```python
import pandas as pd
import numpy as np
```

```python
path_to_data = 'NTSB avall file dataset.xlsx'
ntsb=pd.ExcelFile(path_to_data)
ntsb1 = pd.read_excel(ntsb, 'Aircraft related')
ntsb2 = pd.read_excel(ntsb, 'Event related')
ntsb3 = pd.read_excel(ntsb, 'Flight_Crew')
```

```python
print(len(ntsb1))
print(len(ntsb2))
print(len(ntsb3))
```

```python
ntsb1 = pd.read_excel(path_to_data, sheet_name='Aircraft related', usecols=['ev_id','cert_max_gr_wt','acft_category','homebuilt','afm_hrs_last_insp','damage'])
ntsb2 = pd.read_excel(path_to_data, sheet_name='Event related', usecols=['ev_id','ev_dow', 'ev_time','ev_year', 'ev_month','ev_nr_apt_loc',
       'apt_dist','apt_elev','light_cond', 'sky_cond_nonceil','altimeter',
       'sky_nonceil_ht', 'sky_ceil_ht', 'sky_cond_ceil', 'vis_sm', 'wx_temp', 'wx_dew_pt', 'wind_dir_deg','wind_vel_kts', 'dec_latitude', 'dec_longitude', 'gust_kts','ev_highest_injury','wx_cond_basic','dec_latitude', 'dec_longitude'])
ntsb3 = pd.read_excel(path_to_data, sheet_name='Flight_Crew', usecols=['ev_id','crew_age','med_certf','med_crtf_vldty','crew_tox_perf','crew_no'])
```

```python
dataframes = [ntsb1, ntsb2, ntsb3]
merged_df = dataframes[0]
for df in dataframes[1:]:
    merged_df = pd.merge(merged_df, df, on='ev_id', how='inner')
print(merged_df)
```

```python
merged_df = merged_df[(merged_df['acft_category'] == "AIR")]
print(merged_df)
```

```python
merged_df = merged_df[(merged_df['homebuilt'] == "N")]
print(merged_df)
```

```python
column_names = "', '".join(merged_df.columns)
column_names = f"'{column_names}'"
print(column_names)
```

```python
df_unique = merged_df.drop_duplicates(subset=['ev_id', 'damage'])
print(df_unique)
```

```python
different_damage = df_unique.groupby('ev_id')['damage'].nunique()
ev_ids_with_different_damage = different_damage[different_damage > 1].index
rows_with_different_damage = df_unique[df_unique['ev_id'].isin(ev_ids_with_different_damage)]
print(rows_with_different_damage)
```

```python
df_unique['severity_class'] = df_unique['damage'] + '-' + df_unique['ev_highest_injury']
print(df_unique[['ev_id', 'damage','ev_highest_injury', 'severity_class', 'acft_category','homebuilt','cert_max_gr_wt', 'afm_hrs_last_insp', 'ev_dow',
                 'ev_time', 'ev_year', 'ev_month', 'ev_nr_apt_loc', 'apt_dist', 'apt_elev', 'light_cond', 'sky_cond_nonceil',
                 'sky_nonceil_ht', 'sky_ceil_ht', 'sky_cond_ceil', 'vis_sm', 'wx_temp', 'wx_dew_pt', 'wind_dir_deg', 'wind_vel_kts',
                 'gust_kts', 'altimeter',  'wx_cond_basic', 'dec_latitude', 'dec_longitude', 'crew_no', 'crew_age',
                 'med_certf', 'med_crtf_vldty', 'crew_tox_perf']])
```

```python
cleaned_df = df_unique.dropna(subset=['severity_class'])
print(cleaned_df)
```

```python
cleaned_df_1 = cleaned_df.loc[:, cleaned_df.nunique() > 1]
print(cleaned_df_1)
```

```python
cleaned_df_1.reset_index(drop=True)
```

```python
output_filename = 'data_set.csv'
cleaned_df_1.to_csv(output_filename, index=False)
print(f"Data saved to {output_filename}")
```
