```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data_set.csv")
df.head
```

```python
columns_list = df.columns
count=0
for c in columns_list:
    try:
        plt.hist(df[c])
        plt.title(c)
        plt.show()
        count+=1
    except:
        print(c)
```

```python
sorted_afm_hrs_last_insp = np.sort(df["afm_hrs_last_insp"])
df_outliers = df[df["afm_hrs_last_insp"] > 0]
plt.hist(df_outliers["afm_hrs_last_insp"])
plt.show()
```

```python
df["afm_hrs_last_insp"] = np.where(df["afm_hrs_last_insp"] > 0, df["afm_hrs_last_insp"], np.nan)
```
